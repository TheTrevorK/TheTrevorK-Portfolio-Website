### Drop Unwanted Data

```python
applications.drop(columns= ['FLAG_MOBIL','FLAG_WORK_PHONE','FLAG_PHONE','FLAG_EMAIL','OCCUPATION_TYPE'],inplace=True)
applications.drop_duplicates(subset=applications.columns[1:],inplace=True)
```

```python
applications['EDUCATION_TYPE'] = applications['NAME_EDUCATION_TYPE'].apply(lambda x: x.split()[0].strip() if x else None)
```

```python
applications= applications.drop(columns = ['NAME_EDUCATION_TYPE'])
```
